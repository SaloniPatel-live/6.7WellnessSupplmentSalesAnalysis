### Creating a correlation matrix heatmap using numerical columns in the dataset wellsupp_data.

These numerical columns are
- units_sold
- Price
- Revenue
- Discount
- units_returned

```python
# calculate the correlations

corr_matrix = wellsupp_data.corr(numeric_only=True)
```

```python
# Create a fresh figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)

# Apply layout tweaks
plt.tight_layout()

# Save the figure directly from `fig`
fig.savefig(f"{path}\\Visualizations\\testing2.png", dpi=400, bbox_inches='tight')

# Display in notebook
plt.show()
```

#### Conclusion 

- Price and Revenue has coefficient of 0.98 representating their strong relationship. It means that when the price is high the revenue is increasing.
- units_sold and revenue has coefficient of 0.21 which is not very strong but also not weak either. It is obvious that the more units_sold will convert into higher revenue but probably bulk pricing or tiered pricing mean the revenue is growing at lower pace.
- Discount is very nominal and its coefficient with revenue or units_sold do not seem to have any significant impact. 
    


### Creating Scatter Plots for the strongest correlations (between -0.6 to 0.6 and excluding self correlations like price to price)

```python
# Flatten the matrix and filter pairs with strong correlation
strong_pairs = (
    corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    .stack()
    .reset_index()
    .rename(columns={'level_0': 'Feature_1', 'level_1': 'Feature_2', 0: 'Correlation'})
)
```

```python
# Filter for strong correlations (e.g., > 0.6 or < -0.6)
strong_corrs = strong_pairs[
    strong_pairs['Correlation'].abs() > 0.6
].sort_values(by='Correlation', ascending=False)
```

```python
# Show strongest correlation pairs
print(strong_corrs)
```

#### As suspected there is only one strong correalation - Price and Revenue - in this dataset.

```python
# Create folder if needed
# import os
# os.makedirs(f"{path}\\Visualizations", exist_ok=True)

for i, row in strong_corrs.iterrows():
    x, y = row['Feature_1'], row['Feature_2']
    
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=wellsupp_data, x=x, y=y, hue='Category', palette='Set1')
    plt.title(f"{x} vs {y} (corr={row['Correlation']:.2f})")
    plt.tight_layout()
    plt.savefig(f"{path}\\Visualizations\\scatter_{x}_vs_{y}_category.png", dpi=300)
    plt.show()
```

### Creating a stacked bar chart to show the location breakdown

```python
# Step 1: Aggregate revenue by Category and Location
pivot_df = (
    wellsupp_data.groupby(['Category', 'Location'])['Revenue']
    .sum()
    .reset_index()
    .pivot(index='Category', columns='Location', values='Revenue')
    .fillna(0)
)

# Step 2: Sort Categories by total revenue
pivot_df['Total'] = pivot_df.sum(axis=1)
pivot_df = pivot_df.sort_values('Total', ascending=False)
pivot_df.drop('Total', axis=1, inplace=True)

# Step 3: Plot stacked bars
plt.figure(figsize=(12, 7))
bottom = pd.Series([0]*len(pivot_df), index=pivot_df.index)

for location in pivot_df.columns:
    plt.bar(pivot_df.index, pivot_df[location], bottom=bottom, label=location)
    bottom += pivot_df[location]

# Step 4: Customize chart
plt.title('Stacked Revenue by Category and Location')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Location')
plt.tight_layout()

# Save output
plt.savefig(f"{path}\\Visualizations\\stackedbar_revenue_by_category_location.png", dpi=300)
plt.show()
```

#### Conclusion - Even at the country level Vitamin and Mineral categories are leading in the revenue.

### Creating a line chart for product and prices by quarter

```python
# Creating Quarter column

wellsupp_data['Quarter'] = wellsupp_data['Date'].dt.to_period('Q').astype(str)
```

```python
# Calculate Quarterly Average Price per Product

quarterly_price_trend = (
    wellsupp_data.groupby(['Quarter', 'Product'])['Price']
    .mean()
    .reset_index()
    .rename(columns={'Price': 'AvgPrice'})
)
```

```python
# Creating a pivot for platting

quarterly_pivot = quarterly_price_trend.pivot(index='Quarter', columns='Product', values='AvgPrice').fillna(0)
```

```python
# Creating a line chart

plt.figure(figsize=(12, 7))
for product in quarterly_pivot.columns:
    plt.plot(quarterly_pivot.index, quarterly_pivot[product], marker='o', label=product)

plt.title("Quarterly Average Price Trend by Product")
plt.xlabel("Quarter")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.legend(title="Product")
plt.legend(loc='upper right')  # Try 'upper right', 'lower left', 'center'
plt.tight_layout()
plt.show()

```

#### Conclusion

- Prices are all over the place for quarters for each product.
